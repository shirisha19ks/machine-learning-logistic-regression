### IQR Method

```python
dfsonar.columns
```

```python
for i in range(60):
    print(i)
```

```python
for i in range(60) :
        q1 = dfsonar[i].quantile(0.25)
        q3 = dfsonar[i].quantile(0.75)
        iqr = q3 - q1
        lower = q1 - (1.5 * iqr)
        upper = q3 + (1.5 * iqr)
        dfsonar[i] = dfsonar[i].map(lambda x : lower if x < lower else upper if x > upper else x )
```

```python
plt.figure(figsize = (12, 8))
sns.boxplot(dfsonar)
plt.show()
```

### Auto Random State

```python
sc_list = []
for i in range(500) :
     Xtrain, Xtest, ytrain, ytest = train_test_split(X_ros, y_ros, test_size=0.20, random_state=i)
     logreg_model1 = LogisticRegression()
     logreg_model1.fit(Xtrain, ytrain)
     sc = logreg_model1.score(Xtest, ytest)
     sc_list.append(sc)

m = max(sc_list)
rs = sc_list.index(m)
print(m , "==>", rs)
```
